# Series

```python
data = [10,20,30,40]
series = pd.Series(data)
print(series)
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 45],
        'Salary': [50000, 60000, 70000, 80000, 90000],
        'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales'],
        'Education': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master']}
series = pd.Series(data)
print(series)
```

# DataFrame

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 45],
        'Salary': [50000, 60000, 70000, 80000, 90000],
        'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales'],
        'Education': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master']}
df = pd.DataFrame(data)
print(df)
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Sam', 'Ironman', 'Joy'],
        'Age': [25, 30, 35, 40, 45, 42, 37, 43],
        'Salary': [50000, 60000, 70000, 80000, 90000, 100000, 85000, 95000],
        'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales', 'HR', 'Marketing', "IT"],
        'Education': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master', 'Master', 'Bachelor', 'CS']}
df = pd.DataFrame(data)
print(df)
```

# Combining Multiple Conditions

```python
filter_data = df[(df['Age']>35) & (df['Salary']>50000)]  # combining multiple conditions using and (&), or (|)
```

```python
filter_data
```

```python
filter_data = df[(df['Age']>35) | (df['Salary']>70000)]
```
