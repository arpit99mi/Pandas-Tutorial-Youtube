```python
import pandas as pd
import numpy as np

# Sample sales dataset
data = {
    'region':  ['North', 'South', 'North', 'East', 'South', 'East'],
    'product': ['A', 'B', 'B', 'A', 'A', 'B'],
    'sales':   [200, 150, np.nan, 300, np.nan, 180],
    'units':   [10, 8, 12, 15, 7, 9],
}

df = pd.DataFrame(data)
print(df)
```

# Interpolation

```python
df['sales_linear'] = df['sales'].interpolate(method='linear')
df[['sales', 'sales_linear']]
```

```python
df['sales_polynomial'] = df['sales'].interpolate(method='polynomial', order=2)
df['sales_polynomial']
```

```python
df['sales'].interpolate(method='nearest')
```

# Sorting

```python
sorted_df = df.sort_values(by='units', ascending=False)
sorted_df
```

```python
df_sorted = df.sort_values(by='units', ascending=False)
print(df_sorted[['region', 'product', 'units']])
```

```python
multi_sorted = df.sort_values(by=['region','units'], ascending=[True,False])
print(multi_sorted[['region','units']])
```

```python
df.sort_values(by='sales', na_position='first')
```

```python
df.sort_values(by='sales', na_position='last')
```

```python
shuffled = df.sample(frac=1)
restored = shuffled.sort_index()
restored
```

# Aggregation

```python
print('Sum :', df['units'].sum())
```

```python
print('mean :', df['units'].mean())
```

```python
print('Median :', df['units'].median())
```

# GroupBy

```python
region_total = df.groupby('region')['units'].sum()
region_total
```

```python
grouped = df.groupby(['region', 'product'])['units'].sum()
grouped
```

```python
result = df.groupby('region').agg(
    total_units = ('units', 'sum'),
    avg_sale = ('sales', 'mean'),
    pro_count = ('product', 'count')
)
```
