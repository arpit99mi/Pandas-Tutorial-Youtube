```python
import pandas as pd
import numpy as np
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
        'Age': [22, 25, 23, 21, 24],
        'Score': [85, 78, 92, None, 88],
        'City': ['Delhi', 'Mumbai', 'bangalore', 'Chennai', 'Delhi']}
df = pd.DataFrame(data)
print(df)
```

# Adding Columns

```python
df['Country'] = ['India'] * 5
```

```python
df
```

```python
df['Bonus_Score'] = df['Score'] * 0.1
```

```python
df
```

```python
df['Grade'] = np.where(df['Score'] >=85, 'A', 'B')
```

```python
df
```

# Updating Values

```python
df.loc[1, 'Age'] = 27
```

```python
df
```

```python
df.at[2, 'Age'] = 27
```

```python
df
```

```python
df['Score'] = df['Score'] +5
```

```python
df
```

```python
df['City'] = np.where(df['City'] == 'Delhi', 'NCR', df['City'])
```

```python
df
```

# Removing Rows & Columns

```python
df.drop([1,3], axis=0)
```

```python
df.drop(['Age', 'City'], axis=1)
```

# Handling Missing Values

```python
df.isnull().sum()
```

```python
mean_value = df['Score'].mean()
mean_value
```

```python
df['Score'] = df['Score'].fillna(mean_value)
```

```python
df
```

```python
median_value = df['Score'].median()
median_value
```

```python
df['Score'] = df['Score'].fillna(median_value)
```
